```python
import pandas as pd
import pickle
from sklearn.preprocessing import MinMaxScaler

# 1. Load Dataset
df = pd.read_csv('Data_final.csv')

# 2. Data Cleaning
# Ensure there are no missing values
df = df.dropna()

# 3. Create Career Profiles
# We group by 'Career' and take the mean of all scores.
# This gives us the "Ideal Profile" for each of the 104 careers.
career_profiles = df.groupby('Career').mean().reset_index()

# 4. Scaling (Standardizing scores between 0 and 1)
# This ensures that one trait doesn't dominate others during similarity calculation
features = ['O_score', 'C_score', 'E_score', 'A_score', 'N_score', 
            'Numerical Aptitude', 'Spatial Aptitude', 'Perceptual Aptitude', 
            'Abstract Reasoning', 'Verbal Reasoning']

scaler = MinMaxScaler()
career_profiles[features] = scaler.fit_transform(career_profiles[features])

# 5. Export for Streamlit
pickle.dump(career_profiles, open('career_profiles.pkl', 'wb'))
pickle.dump(scaler, open('scaler.pkl', 'wb'))

print("Career profiles exported successfully!")
```
